# Inspect frozen Behavior-to-Weights results
This notebook reads outputs only; it does not perform confirmatory inference.


```python
from pathlib import Path
import json
import pandas as pd
RESULTS = Path('../artifacts/results.jsonl')
rows = [json.loads(line) for line in RESULTS.open() if line.strip()]
frame = pd.DataFrame(rows)
frame.head()
```

```python
summary = (frame.groupby(['split','channel','query_budget','method','metric'])['value']
           .agg(['count','mean','std']).reset_index())
summary
```
